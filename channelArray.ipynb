```python
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
import itertools
from dataPreprocess import read_dataset
import h5py

sns.set_theme(style='whitegrid', font_scale=2)

nwb = h5py.File('/home/jovyan/dataset/indy/The_nwb_Raw_Dataset/indy_20160622_01.nwb', mode='r')
array_map = (nwb['/general/extracellular_ephys/electrode_map'][()][:, :2] / 0.0004).astype(int)

array_map = array_map - np.min(array_map, axis=0)

arr = np.full([10, 10], fill_value=np.nan)
for n, (ax, ay) in enumerate(array_map):
    arr[ax, ay] = n+1

plt.figure(figsize=(15, 15))
sns.heatmap(arr, cbar=False, xticklabels=False, yticklabels=False, square=True, \
    annot=True, cmap=sns.dark_palette("#69d", reverse=True, as_cmap=True))
```
